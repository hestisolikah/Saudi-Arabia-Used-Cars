**Metric Evaluation**

Penghapusan data outlier pada model ini hanya dilakukan oleh data outlier dengan nilai ekstrem yang tidak realistis. Oleh karena itu, data masih akan memiliki beberapa outlier yang tertinggal. Hal ini membuat metrik evaluasi model yang akan digunakan tidak sensitif terhadap outlier. Sehingga, beberapa metrik yang digunakan dalam model ini diantaranya adalah:

1. MAE (Mean Absolute Error): Metrik ini menggambarkan selisih rata-rata absolut antara hasil prediksi dan hasil aktual. Metrik ini direkomendasikan karena data yang digunakan memiliki beberapa outlier. Karena menggunakan nilai absolut, metrik ini tidak peduli dengan selisih yang menghasilkan hasil negatif. Semakin kecil nilai metrik ini, semakin baik hasil prediksi yang diperoleh. 

2. MAPE (Mean Absolute Percentage Error): Metrik ini menggambarkan persentase rata-rata selisih absolut antara hasil prediksi dan hasil aktual. MAPE memiliki perbedaan dengan MAE, yaitu metrik ini perlu mengubah selisih menjadi persentase saja. Karena merupakan persentase, nilai metrik ini berkisar antara 0 hingga 1.

3. RMSLE(Root Mean Squared Log Error): Metrik ini digunakan saat data memiliki skewness yang tinggi dengan variasi yang berkisar dari kecil hingga sangat besar. Perbedaan varians yang besar dapat diatasi karena perhitungan RMSLE adalah dengan mengonversi data ke skala logaritmik terlebih dahulu, kemudian dihitung rata-ratanya dan diakar pangkat dua. Dataset yang memenuhi persyaratan tersebut sehingga metrik ini cocok digunakan untuk mengevaluasi hasil dari model.

4. R- square: Metrik ini menggambarkan persentase sejauh mana semua fitur mempengaruhi target. Semakin tinggi nilai persentasenya, semakin besar pengaruh feature terhadap target. Metrik ini hanya digunakan saat memilih model dasar karena dari setiap model dapat terlihat dengan jelas seberapa besar persentase pengaruh feature terhadap target.

Metrik seperti MSE(Mean Squared Error), RMSE(Root Mean Squared Error), dan RMPSE(Root Mean Square Percentage Error) tidak cocok digunakan dalam data ini karena sensitif terhadap adanya outlier. Outlier merujuk pada data yang jauh berbeda atau ekstrim dari sebagian besar data lainnya. Outlier dapat memiliki dampak yang signifikan pada hasil metrik seperti MSE, RMSE, dan RMSPE karena metrik-metrik tersebut menghitung perbedaan kuadrat atau persentase perbedaan antara nilai aktual dan nilai prediksi. Outlier dengan nilai yang ekstrem dapat memberikan kontribusi yang tidak proporsional terhadap perhitungan metrik, sehingga hasil evaluasi menjadi bias dan tidak akurat. 

Dalam konteks ini, metrik evaluasi yang lebih tahan terhadap outlier, seperti MAE dapat lebih sesuai untuk digunakan. Metrik-metrik ini mengukur perbedaan absolut antara nilai aktual dan nilai prediksi, sehingga lebih stabil terhadap keberadaan outlier dalam data. Dengan menggunakan metrik evaluasi yang lebih tahan terhadap outlier, evaluasi kualitas prediksi model akan lebih akurat dan dapat memberikan informasi yang lebih relevan.

**Extreme Outliers**

Dalam analisis selanjutnya, kita perlu menghapus indeks yang tidak diperlukan berdasarkan pengetahuan domain yang kita miliki. Dalam menganalisis data, tentu saja kita akan berpikir bahwa adanya data outlier sangat mengganggu. Namun, tidak semua data outlier harus dihapus karena penempatan data tersebut benar sesuai dengan fakta dan realitas yang diterima. Dalam ilmu data, seringkali kita membuat asumsi tentang populasi tertentu. Namun, adanya nilai outlier ekstrem dapat memiliki dampak signifikan pada hasil interpretasi yang kita amati dari data atau model machine learning. Dengan penanganan yang tepat, kita dapat mempertimbangkan data outlier ekstrem sebagai populasi yang berbeda.

Sebelum menghadapi data outlier ekstrem, mari kita coba melihat bagian indeks mana yang tidak masuk akal.

Dalam konteks ini, kita akan mencari indeks yang tidak memiliki relevansi atau tidak memberikan kontribusi signifikan dalam analisis yang kita lakukan. Dengan menghapus indeks yang tidak berguna, kita dapat menyederhanakan data dan fokus pada variabel-variabel yang lebih penting untuk membangun model prediktif yang akurat.

```python
def check_outlier(col):
    
    Q1 = np.percentile(col, 25)
    Q3 = np.percentile(col, 75)

    IQR = Q3 - Q1

    batas_bawah = Q1 - (IQR * 1.5)
    batas_atas = Q3 + (IQR * 1.5)

    data_outliers = col[(col < batas_bawah) | (col > batas_atas)]
    return data_outliers
```

```python
feature_outlier = []
for col in df.select_dtypes(['float','int']):
    outliers = check_outlier(df[col])
    if len(outliers):
        print(f"Feature",col, "has", len(outliers), "outlier")
        feature_outlier.append(col)
    else:
        print(f"Feature", col, "has no outlier")
```

Multicollinearity adalah fenomena di mana dua atau lebih fitur dalam model regresi memiliki korelasi tinggi antara satu sama lain. Hal ini dapat menyebabkan masalah dalam interpretasi hasil model, karena dapat sulit untuk menentukan kontribusi individu dari masing-masing fitur terhadap variabel target. Dalam konteks ini, kita perlu memeriksa apakah ada korelasi tinggi antara fitur-fitur lain dalam dataset yang dapat mempengaruhi interpretasi model.

Untuk memeriksa multicollinearity, kita dapat menggunakan matriks korelasi antara fitur-fitur yang ada dalam dataset. Jika terdapat korelasi yang tinggi antara dua atau lebih fitur, hal itu menunjukkan adanya indikasi multicollinearity.

Selain itu, kita juga dapat menggunakan faktor inflasi varians (VIF) untuk mengukur tingkat multicollinearity antara fitur-fitur. Jika VIF suatu fitur melebihi batas tertentu, itu menunjukkan adanya multicollinearity yang signifikan dengan fitur lain dalam model.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif['variables'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return(vif)
```

```python
calc_vif(df[feature_outlier])
```

Pada teori yang ada biasanya kita menganggap nilai VIF yang lebih dari 4 sebagai indikasi multicollinearity. Dalam kasus ini, jika ada fitur dengan nilai VIF lebih dari 4, kita diharapkan untuk menghapus fitur tersebut dari model karena kemungkinan besar fitur tersebut memiliki korelasi tinggi dengan fitur lain dalam dataset.

Namun, dalam dataset ini, kita sebaiknya tidak menghapus fitur-fitur yang memiliki nilai VIF lebih dari 4, yaitu "Year" dan "Engine_Size", meskipun ada indikasi multicollinearity. Hal ini dikarenakan fitur-fitur tersebut memiliki pengaruh besar terhadap penjualan mobil bekas.

Dalam konteks ini, "Year" dan "Engine_Size" mungkin memberikan informasi penting tentang nilai mobil bekas dan preferensi konsumen terhadap ukuran mesin. Oleh karena itu, meskipun terdapat multicollinearity, kita sebaiknya mempertahankan fitur-fitur tersebut dalam model.

Namun, penting untuk diingat bahwa adanya multicollinearity dapat mempengaruhi interpretasi hasil model. Oleh karena itu, saat menganalisis hasil model, kita perlu berhati-hati dalam menafsirkan kontribusi individu dari masing-masing fitur yang terkait dengan multicollinearity.

**Encoding**

Untuk dapat menggunakan semua fitur yang dipilih dalam pemodelan, fitur-fitur kategorikal harus diubah menjadi fitur numerikal. Hal ini dapat dilakukan dengan menggunakan ColumnTransformer, yang merupakan salah satu komponen dalam pipeline model.

ColumnTransformer adalah sebuah alat dalam pemrosesan data yang memungkinkan kita untuk menerapkan transformasi yang berbeda pada setiap kolom data. Dalam konteks ini, kita dapat menggunakan ColumnTransformer untuk mengubah tipe data fitur-fitur kategorikal menjadi fitur numerikal. Transformasi ini dapat dilakukan dengan menggunakan teknik seperti One-Hot Encoding atau Label Encoding, tergantung pada jenis data dan konteks fitur yang sedang kita proses.

Setelah kita mendefinisikan ColumnTransformer dengan transformasi yang tepat untuk setiap kolom, kita dapat mengintegrasikannya ke dalam pipeline model. Pipeline model adalah serangkaian langkah pemrosesan data yang terdiri dari transformasi dan estimasi (model). Dalam kasus ini, ColumnTransformer akan menjadi salah satu langkah dalam pipeline, yang akan mengubah tipe data fitur-fitur kategorikal menjadi numerikal sebelum dilakukan proses pemodelan.

Dengan menggunakan ColumnTransformer dan pipeline model, kita dapat secara efisien mengubah tipe data fitur-fitur kategorikal menjadi numerikal, sehingga memungkinkan kita untuk menggunakan semua fitur yang telah dipilih dalam pemodelan.

```python
listdesc = []
for col in df.select_dtypes('object').columns:
    listdesc.append([col,
                     df[col].dtypes,
                     df[col].nunique(),
                     df[col].drop_duplicates().sample(1).values])

pd.DataFrame(listdesc, columns=['Data Features',
                               'Data Types',
                               'Unique',
                               'Unique Sample'])
```

```python
ct = ColumnTransformer([
    ('One Hot Encoder', OneHotEncoder(drop='first'), ['Gear_Type', 'Options']),
    ('Binary Encoder', BinaryEncoder(), ['Type', 'Region', 'Make'])
], remainder='passthrough')
```

Syntax di atas adalah penggunaan `ColumnTransformer` dari modul `sklearn.compose` dalam scikit-learn library untuk melakukan transformasi kolom pada dataset.

Pada syntax tersebut, kita membuat objek `ColumnTransformer` dengan nama `ct`. `ColumnTransformer` digunakan untuk menggabungkan beberapa transformer yang berbeda pada kolom yang berbeda dalam dataset yang sama.

Penjelasan untuk setiap argumen yang digunakan dalam `ColumnTransformer`:

1. `'One Hot Encoder'`: Nama untuk transformer pertama yang akan digunakan.
2. `OneHotEncoder(drop='first')`: Objek transformer pertama yang digunakan. Di sini kita menggunakan `OneHotEncoder` dari modul `sklearn.preprocessing` untuk melakukan encoding pada fitur-fitur `'Gear_Type'` dan `'Options'`. Argumen `drop='first'` digunakan untuk menghapus kolom pertama dari setiap fitur yang di-encode untuk menghindari jebakan dummy variable.
3. `['Gear_Type', 'Options']`: Daftar fitur yang akan di-encode menggunakan `OneHotEncoder`.
4. `'Binary Encoder'`: Nama untuk transformer kedua yang akan digunakan.
5. `BinaryEncoder()`: Objek transformer kedua yang digunakan. Di sini kita menggunakan `BinaryEncoder` dari library `category_encoders` untuk melakukan encoding pada fitur-fitur `'Type'`, `'Region'`, dan `'Make'`.
6. `['Type', 'Region', 'Make']`: Daftar fitur yang akan di-encode menggunakan `BinaryEncoder`.
7. `remainder='passthrough'`: Argumen ini digunakan untuk menentukan apa yang harus dilakukan pada kolom yang tidak mengalami transformasi. `'passthrough'` berarti kolom-kolom yang tidak di-encode atau di-transformasikan akan dilewatkan (tidak diubah) ke output tanpa perubahan.

Dengan menggunakan `ColumnTransformer`, kita dapat menggabungkan beberapa transformer yang berbeda dalam satu proses dan mengaplikasikan transformasi ini pada dataset dengan mudah. Setelah membuat objek `ColumnTransformer`, Anda dapat menggunakan `fit_transform()` atau `transform()` untuk menerapkan transformasi pada dataset yang sebenarnya.

```python

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
```

**Choose a Benchmark Model**

```python
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=100)
rf = RandomForestRegressor(random_state=100)
gdt = GradientBoostingRegressor(random_state=100)
xgb = XGBRegressor(random_state=100)
ridge = Ridge()
lasso = Lasso()
elastic_net = ElasticNet()
Ada_Boost = AdaBoostRegressor()
lgbm = LGBMRegressor(random_state=100)
cat = CatBoostRegressor(random_state=100, verbose=False)


ttr_lr = TransformedTargetRegressor(lr, func = np.log, inverse_func = np.exp)
ttr_knn = TransformedTargetRegressor(knn, func = np.log, inverse_func = np.exp)
ttr_dt = TransformedTargetRegressor(dt, func = np.log, inverse_func = np.exp)
ttr_rf = TransformedTargetRegressor(rf, func = np.log, inverse_func = np.exp)
ttr_gdt = TransformedTargetRegressor(gdt, func = np.log, inverse_func = np.exp)
ttr_xgb = TransformedTargetRegressor(xgb, func = np.log, inverse_func = np.exp)
ttr_ridge = TransformedTargetRegressor(ridge, func = np.log, inverse_func = np.exp)
ttr_lasso = TransformedTargetRegressor(lasso, func = np.log, inverse_func = np.exp)
ttr_enet = TransformedTargetRegressor(elastic_net, func = np.log, inverse_func = np.exp)
ttr_Ada_Boost = TransformedTargetRegressor(Ada_Boost, func = np.log, inverse_func = np.exp)
ttr_lgbm = TransformedTargetRegressor(lgbm, func = np.log, inverse_func = np.exp)
ttr_cat = TransformedTargetRegressor(cat, func = np.log, inverse_func = np.exp)


models = [ttr_lr, ttr_knn, ttr_dt, ttr_rf, ttr_gdt, ttr_xgb,  ttr_ridge, ttr_lasso, ttr_enet, ttr_Ada_Boost, ttr_lgbm, ttr_cat]


score_mae = []
score_mape = []
score_rmsle = []
score_r2 = []

for model in models:
    
    cross_val = KFold(n_splits = 5, shuffle = True, random_state = 100)
    estimator = Pipeline([
        ('preprocessing', ct),
        ('model', model)
    ])

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_absolute_error', 
        error_score = 'raise'
        )

    score_mae.append(model_cv_mae.mean())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_absolute_percentage_error', 
        error_score = 'raise'
        )

    score_mape.append(model_cv_mape.mean())
    
       # RMSLE
    model_cv_rmsle = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'neg_mean_squared_log_error', 
        error_score = 'raise'
        )

    score_rmsle.append(model_cv_rmsle.mean())

       # R2
    model_cv_r2  =  cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv = cross_val, 
        scoring = 'r2', 
        error_score = 'raise'
        )

    score_r2.append(model_cv_r2.mean())

score_rmsle = np.sqrt(abs(np.array(score_rmsle)))
```

- Catboost menghasilkan hasil canggih tanpa pelatihan data ekstensif yang biasanya diperlukan oleh metode pembelajaran mesin lainnya.
- Memberikan dukungan out-of-the-box yang kuat untuk format data yang lebih deskriptif yang menyertai banyak masalah bisnis.

**Hyperparameter Tuning**

Pada tahap ini, kita akan melakukan tuning hyperparameter untuk model Cat Boost. Hyperparameter tuning adalah proses mencari kombinasi parameter terbaik untuk meningkatkan kinerja model. Hal ini dilakukan dengan mencoba berbagai kombinasi parameter dan memilih yang memberikan hasil terbaik. Dalam kasus ini, kita akan mencari kombinasi terbaik untuk model Cat Boost dengan menggunakan grid search atau random search. Sebelumnya, saya telah melakukan percobaan dengan random search namun hasilnya pada model setelah dilakukan tuning memiliki nilai yang lebih baik dengan penggunaan grid search. Proses ini diharapkan akan membantu kita meningkatkan performa model dan mendapatkan hasil prediksi yang lebih akurat.

```python
hyperparam_space_catboost = {
    'model__learning_rate': [0.01, 0.1, 0.5],
    'model__iterations': [175, 225, 275],
    'model__depth': [i for i in range(7,11,1)],
    'model__l2_leaf_reg': [i for i in range(4,7,1)]
}
```

```python
hyperparam_space_catboost = {'model__depth': [9],
 'model__iterations':[275],
 'model__l2_leaf_reg': [4],
 'model__learning_rate': [0.1]}
```

```python
catboost = CatBoostRegressor(random_state=100,verbose=False)

crossval = KFold(n_splits=5, shuffle=True, random_state=100)

estimator_catboost = Pipeline([
    ('preprocessing', ct),
    ('model', catboost)
])

gs = GridSearchCV(
estimator_catboost,
param_grid=hyperparam_space_catboost,
cv=crossval,
scoring=['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'],
refit='neg_mean_absolute_error',
n_jobs=-1
)
```

```python
gs.fit(X_train, y_train)
```

```python
print('Cat Boost')
print('Best_score:', gs.best_score_)
print('Best_params:', gs.best_params_)
```

Hasil dari penelusuran parameter menggunakan GridSearchCV menunjukkan bahwa parameter terbaik untuk model CatBoostRegressor adalah sebagai berikut:

'depth': 9: Parameter ini mengontrol kedalaman maksimum dari pohon yang digunakan dalam model. Dalam kasus ini, kedalaman maksimum pohon adalah 9. 

'iterations': 275: Parameter ini menentukan jumlah iterasi atau langkah yang akan dilakukan dalam proses gradient boosting. Model akan mempelajari pola data dalam sebanyak 275 iterasi.

'l2_leaf_reg': 4: Parameter ini merupakan regulasi L2 yang digunakan dalam model untuk mencegah overfitting. Nilai 4 menunjukkan kekuatan regulasi yang diterapkan pada model.

'learning_rate': 0.1: Parameter ini menentukan tingkat pembelajaran atau seberapa cepat model belajar dari data. Dalam kasus ini, tingkat pembelajaran adalah 0.1.
